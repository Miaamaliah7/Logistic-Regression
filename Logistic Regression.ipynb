```python
x=pd.DataFrame(dbt.iloc[:,:-1])
y=pd.DataFrame(dbt.iloc[:,-1])
print (x)
```

```python
print (y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
dbtmodel=LogisticRegression()
dbtmodel.fit(x_train,y_train)
```

```python
y_pred=dbtmodel.predict(x_test)
y_pred
```

```python
print("Accuracy",(dbtmodel.score(x_test,y_test)))
```

```python
confusion_matrix=confusion_matrix(y_test,y_pred)
print(confusion_matrix)
```
